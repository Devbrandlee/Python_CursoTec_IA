**Mini-Desafio: Calculadora Simples**
>Crie uma calculadora que:

```
• Receba dois números do usuário
• Realize as 4 operações básicas (+, -, *, /)
• Exiba os resultados

```





````python
num1 = float(input("Digite o primeiro número"))
num2 = float(input("Digite o segundo número"))

soma = num1 + num2

divisao = num1 / num2

sub = num1 - num2

mult = num1 * num2


print("O Resultado da Soma é:", soma)

print("O resultado da Divisão é:", divisao)

print("O resultado da subtração é:", sub)

print("O resultado da multplicação é:", mult)
````
