```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder

df = pd.read_csv('dados[1].csv')
df.columns = df.columns.str.strip()
df = df.dropna(subset=['Resultado'])

colunas_notas = ['A_NotaRegular', 'A_NotaRecuperacao', 'B_NotaRegular', 'B_NotaRecuperacao', 'C_NotaRegular', 'C_NotaRecuperacao', 'NotaTotal']
for col in colunas_notas:
    df[col] = df[col].astype(str).str.replace('-', '0').str.strip()
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

colunas_acesso = [
    'Página Geral', 'Fórum Geral', 'Arquivo Conteudo', 'Relatório', 'URL Geral',
    'Arquivo A', 'Tarefa A', 'URL A', 'Fórum A', 'Página A', 'Pasta A', 'Questionário A',
    'Arquivo B', 'Tarefa B', 'URL B', 'Fórum B', 'Pasta B', 'Questionário B', 'Chat B',
    'Arquivo C', 'Tarefa C', 'URL C', 'Fórum C', 'Questionário C', 'Chat C'
]
df[colunas_acesso] = df[colunas_acesso].fillna(0)

label_encoder = LabelEncoder()
df['Resultado_Encoded'] = label_encoder.fit_transform(df['Resultado'])

scaler = StandardScaler()
X_normalizado = scaler.fit_transform(df[colunas_acesso])
df_acessos_normalizados = pd.DataFrame(X_normalizado, columns=colunas_acesso)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('dados[1].csv')
df = df.dropna(subset=['Resultado'])

cols_nota = ['A_NotaRegular', 'A_NotaRecuperacao', 'B_NotaRegular', 'B_NotaRecuperacao', 'C_NotaRegular', 'C_NotaRecuperacao']
for col in cols_nota:
    df[col] = df[col].astype(str).str.replace('-', '0').str.strip()
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

df['A_Final'] = df[['A_NotaRegular', 'A_NotaRecuperacao']].max(axis=1)
df['B_Final'] = df[['B_NotaRegular', 'B_NotaRecuperacao']].max(axis=1)
df['C_Final'] = df[['C_NotaRegular', 'C_NotaRecuperacao']].max(axis=1)

df['A_Status'] = np.where(df['A_Final'] >= 60, 'APROVADO', 'REPROVADO')
df['B_Status'] = np.where(df['B_Final'] >= 60, 'APROVADO', 'REPROVADO')
df['C_Status'] = np.where(df['C_Final'] >= 60, 'APROVADO', 'REPROVADO')

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
disciplinas = ['A', 'B', 'C']
status_cols = ['A_Status', 'B_Status', 'C_Status']
cores = ['skyblue', 'salmon', 'lightgreen']

for i, (disc, col) in enumerate(zip(disciplinas, status_cols)):
    counts = df[col].value_counts().reindex(['APROVADO', 'REPROVADO']).fillna(0)
    bars = axes[i].bar(counts.index, counts.values, color=cores[i], edgecolor='black')
    axes[i].set_title(f'Resultado na Disciplina {disc}')
    axes[i].set_xlabel('Resultado')
    axes[i].set_ylabel('Quantidade de Alunos')
    axes[i].set_ylim(0, max(counts.values) * 1.15)

    for bar in bars:
        height = bar.get_height()
        axes[i].annotate(f'{int(height)}',
                         xy=(bar.get_x() + bar.get_width() / 2, height),
                         xytext=(0, 3),
                         textcoords="offset points",
                         ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Lendo o arquivo real
df = pd.read_csv('dados[1].csv')

# 2. Removendo linhas inválidas
df = df.dropna(subset=['Resultado'])

# 3. Limpando os caracteres '-' das colunas de notas e convertendo para número
cols_nota = ['A_NotaRegular', 'A_NotaRecuperacao', 'B_NotaRegular', 'B_NotaRecuperacao', 'C_NotaRegular', 'C_NotaRecuperacao']
for col in cols_nota:
    df[col] = df[col].astype(str).str.replace('-', '0').str.strip()
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

# 4. Calculando a nota final considerando a recuperação
df['A_Final'] = df[['A_NotaRegular', 'A_NotaRecuperacao']].max(axis=1)
df['B_Final'] = df[['B_NotaRegular', 'B_NotaRecuperacao']].max(axis=1)
df['C_Final'] = df[['C_NotaRegular', 'C_NotaRecuperacao']].max(axis=1)

# 5. Criando a variável df_aprovados (AQUI SE RESOLVE O ERRO DA SUA IMAGEM)
df_aprovados = df[df['Resultado'] == 'APROVADO']

# 6. Gerando e estilizando os gráficos de caixas (Box Plot)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
notas_cols = ['A_Final', 'B_Final', 'C_Final']
disciplinas = ['A', 'B', 'C']
cores = ['lightblue', 'lightpink', 'lightgreen']

for i, (col, cor) in enumerate(zip(notas_cols, cores)):
    axes[i].boxplot(df_aprovados[col], patch_artist=True,
                    boxprops=dict(facecolor=cor, color='black'),
                    medianprops=dict(color='red', linewidth=1.5))
    axes[i].set_title(f'Notas da Disciplina {disciplinas[i]} (Aprovados)')
    axes[i].set_ylabel('Nota Final')
    axes[i].set_xticklabels([f'Disciplina {disciplinas[i]}'])
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
