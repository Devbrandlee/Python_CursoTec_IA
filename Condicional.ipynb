```python
#Aula17 unidade 03

dia = input("Digite um dia da semana: ").lower()

dias_validos = ["segunda", "terça", "quarta", "quinta", "sexta", "sábado", "domingo"]

if dia not in dias_validos:
    print("Opção inválida")
else:
    match dia:
        case "segunda" | "terça":
            print("Início da semana")
        case "sexta":
            print("Sextou!")
        case _:  # aqui ainda funciona como "meio da semana"
            print("Meio da semana")

''' O .lower() serve para transformar todo o texto digitado pelo usuário em letras minúsculas.
Isso evita problemas de comparação quando alguém digita com maiúsculas ou misturado. '''
```
