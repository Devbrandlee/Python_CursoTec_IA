```python
#Range gera sequências númericas para usar com for.

"""for i in range (5):
  print(i)"""

for i in range (1,6):
    print(i)
```

```python
#Range(0, 10, 2) 0 parametro do início ; 10 fim ; 2 passo ( ou seja de 2 em dois)
for i in range (0,10,2):
  print(i)
```

```python
#Range (10,0,-1) 10 início ; 0 fim ; -1 passo (Ordem decrescente)

for i in range (10,0,-1):
  print(i)
```

```python
#Laço while - Repetição condicionalaula 27

'''
while condição:

  codigo a repetir
  atualizar condição

'''

contador = 0
while contador < 5:
  print(contador)

  contador = contador + 1
```

Aulas práticas

```python
#Depois do for sempre virá uma variável (que recebe o valor dos elemntos 0 1 2 3 4) e o range gerá a quantidade de valores

for numero_de_execucao in range (5):
  print(f" Execução número {numero_de_execucao}: Olá!")



```

```python
#Depois do for sempre virá uma variável (que recebe o valor dos elemntos e n a posição:0,1,2) e o range gerá a quantidade de valores

for numero_de_execucao in range (1,4): #1,2,3
  print(f" Execução número {numero_de_execucao}: Olá!")
```

```python
#While

contador = 0
while (contador < 22) : #Enquanto for verdadeiro
  print(contador)
  contador = contador + 3
print("Comando fora do bloco!")
```
