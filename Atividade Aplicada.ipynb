```python
#Solução do desafio proposto:


import numpy as np
import matplotlib.pyplot as plt
import csv

dados = []
nome_arquivo = 'fashion_mnist_30_balanceado.csv'

try:
    with open(nome_arquivo, 'r', encoding='utf-8') as meuarquivo:
        leitor = csv.reader(meuarquivo)
        next(leitor) # Pula
        dados = list(leitor)
    print("Arquivo carregado com sucesso!")
except FileNotFoundError:
    print(f"Erro: Arquivo '{fashion_mnist_30_balanceado.csv}' não encontrado.")

classes = {
    0: "Camiseta", 1: "Calça", 2: "Suéter", 3: "Vestido", 4: "Casaco",
    5: "Sandália", 6: "Camisa", 7: "Tênis", 8: "Bolsa", 9: "Bota"
}

linha = 16

label = int(dados[linha][0])
nome_classe = classes[label]

pixels = [int(p) for p in dados[linha][1:]]
imagem = np.array(pixels).reshape(28, 28)

plt.imshow(imagem, cmap='gray')
plt.title(f"Linha {linha} - {nome_classe}")
plt.axis('off')
plt.show()

print(f"A classe identificada é: {nome_classe}")
```
