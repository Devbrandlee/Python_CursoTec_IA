```python
#Operadores Relacionais

"""
== Igual a
!= Diferente de
> Maior que
>= Maior ou igual a
< Menor que
<= Menor ou igual a

"""
a = 2
b = 3
c = 3

#igual A

resultado = ( a == b)
print( "A afirmação é: " , resultado)


#Diferente de

resultado = ( a != b)
print(resultado)

# > Maior que
resultado = (a > c)
print(resultado)

# >= Maior ou igual a
resultado = (b >= c)
print(resultado)

# < Menor que
resultado = ( a < c)
print(resultado)


# <= Menor ou igual a
resultado = (c <= a)
print(resultado)
```
