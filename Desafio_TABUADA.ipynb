```python
# Solicita ao usuário o número desejado

numero =  int(input("Digite um número para visualizar a sua tabuada: "))
print()
contador = 0
while contador <=10:
  resultado = numero * contador
  print(f"{numero} x {contador} = {resultado}")
  contador = contador + 1
print()
print("Tabuada concluída!")
```
